Обучим без нормализации и стандартизации

```python
from sklearn.datasets import load_wine
X, y = load_wine(return_X_y=True)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3, random_state=42)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier()
clf = model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
y_pred = clf.predict(X_test)
```

```python
roc_auc_score(y_test, clf.predict_proba(X_test), multi_class='ovr')
```

Сделаем нормализацию

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1.5))
scaler.fit(X)
normalized_X = scaler.transform(X)
print(normalized_X)
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(normalized_X, y, train_size=0.3, random_state=42)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier()
clf = model.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
roc_auc_score(y_test, clf.predict_proba(X_test), multi_class='ovr')
```
